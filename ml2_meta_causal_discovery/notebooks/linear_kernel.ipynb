```python
import h5py

data_file = "/vol/bitbucket/ad6013/Research/CausalStructureNeuralProcess/ml2_meta_causal_discovery/datasets/data/synth_training_data/linear_gaussian/train_data.hdf5"

with open(data_file, 'rb') as f:
    data = h5py.File(f, 'r')
    X = data['data'][:]
    Y = data['label'][:]
```

```python
import matplotlib.pyplot as plt

for i in range(20):
    print(Y[i])
    print(X.mean(1), X.std(1))
    plt.scatter(X[i, :, 0], X[i, :, 1], cmap='viridis')
    plt.title(f"Variable")
    plt.show()
    plt.close()
```

```python
from ml2_meta_causal_discovery.datasets.dataset_generators import \
    DatasetGenerator, ClassifyDatasetGenerator


dataset_generator = ClassifyDatasetGenerator(
    num_variables=3,
    function_generator="gplvm",
    batch_size=1,
    num_samples=2000,
    kernel_sum=True,
    mean_function="latent",
    graph_type="ER",
    graph_degrees=3,
)
```

```python
import matplotlib.pyplot as plt
for i in range(10):
    (
        target_data,
        causal_graphs,
    ) = next(dataset_generator.generate_next_dataset())
    print(causal_graphs)
    plt.scatter(target_data[0, :, 0], target_data[0, :, 1])
    plt.show()
    plt.close()
```
